```python
class ListNode:
    def __init__(self, val):
        self.val: int = val
        self.next: ListNode | None = None
```

```python
class LinkedListStack:
    def __init__(self):
        self._peek: ListNode | None = None
        self._size: int = 0

    def size(self) -> int:
        return self._size
    
    def is_empty(self) -> bool:
        return self._size == 0
    
    def push(self, val: int):
        node = ListNode(val)
        node.next = self._peek
        self._peek = node
        self._size += 1

    def pop(self) -> int:
        num = self.peek()
        self._peek = self._peek.next
        self._size -= 1

    def peek(self) -> int:
        if self.is_empty:
            raise IndexError('栈为空')
        return self._peek.val
    
    def to_list(self) -> list[int]:
        arr = []
        node = self._peek
        while node:
            arr.append(node.val)
            node = node.next
        return arr[: : -1]
```

```python
class ArrayStack:
    def __init__(self):
        self._stack: list[int] = []

    def size(self) -> int:
        return len(self._stack)
    
    def is_empty(self) -> bool:
        return self.size() == 0
    
    def push(self, val: int):
        self._stack.append(val)

    def pop(self) -> int:
        if self.is_empty:
            raise IndexError('栈为空')
        return self._stack.pop()
    
    def peek(self) -> int:
        if self.is_empty:
            raise IndexError('栈为空')
        return self._stack[-1]
    
    def to_list(self) -> list[int]:
        return self._stack
```

```python
class LinkedListQueue:

    def __init__(self):
        self._front: ListNode | None = None
        self._rear: ListNode | None = None
        self._size: int = 0

    def size(self) -> int:
        return self._size
    
    def is_empty(self) -> bool:
        return self.size() == 0
    
    def push(self, val: int):
        node = ListNode(val)
        if self._front is None:
            self._front = node
            self._rear = node
        else:
            self._rear.next = node
            self._rear = node
        self._size += 1

    def pop(self) -> int:
        num = self.peek()

        self._front = self._front.next
        self._size -= 1
        return num
    
    def peek(self) -> int:
        if self.is_empty:
            raise IndexError('queue is empty.')
        
        return self._front.val
    
    def to_list(self) -> list:
        queue = []
        temp = self._front
        while temp:
            queue.append(temp.val)
            temp = temp.next
        return queue
```

```python
class ArrayQueue:

    def __init__(self, size: int):
        self._nums: list[int] = [0] * size
        self._front: int = 0
        self._size: int = 0

    def capacity(self) -> int:
        return len(self._nums)
    
    def size(self) -> int:
        return self._size
    
    def is_empty(self) -> bool:
        return self._size == 0
    
    def push(self, num: int):
        if self._size == self.capacity():
            raise IndexError('queue is full.')
        rear: int = (self._front + self._size) % self.capacity()
        self._nums[rear] = num
        self._size += 1

    def pop(self) -> int:
        num = self.peek()
        self._front = (self._front + 1) % self.capacity()

        self._size -= 1
        return num

    def peek(self) -> int:
        if self.is_empty():
            raise IndexError('queue is empty.')
        
        return self._nums[self._front]
    
    def to_list(self) -> list[int]:
        res = [0] * self.size()
        j = self._front
        for i in range(self.size):
            res[i] = self._nums[(j % self.capacity())]
            j += 1
        return res
```

```python
from collections import deque

# initialize bidirectional queue
deq: deque[int] = deque()

# element joining the team
deq.append(1)
deq.append(3)
deq.append(5)
deq.append(2)
deq.append(4)

# accessing element
front: int = deq[0]
rear: int = deq[-1]

# element out of the team
pop_front: int = deq.popleft()
pop_rear: int = deq.pop()

# get the length of the bidirectional queue
size: int = len(deq)

# determine if the bidirectional queue is empty
is_empty: bool = len(deq) == 0
```

```python
class ListNode:

    def __init__(self, val: int):
        self.val: int = val
        self.next: ListNode | None = None
        self.pre: ListNode | None = None


class LinkedListDeque:

    def __init__(self):
        self._front: ListNode | None = None
        self._rear: ListNode | None = None
        self._size: int = 0

    def size(self) -> int:
        return self._size
    
    def is_empty(self) -> bool:
        return self.size() == 0
    
    def push(self, num: int, is_front: bool):
        node = ListNode(num)

        if self.is_empty():
            self._front = self._rear = node

        elif is_front:
            self._front.pre = node
            node.next = self._front
            self._front = node

        else:
            self._rear.next = node
            node.pre = self._rear
            self._rear = node

        self._size += 1

    def push_first(self, num: int):
        self.push(num, True)

    def push_last(self, num: int):
        self.push(num, False)

    def pop(self, is_front: bool) -> int:
        if self.is_empty():
            raise IndexError('out of index')
        
        if is_front:
            val: int = self._front.val
            fnext: ListNode | None = self._front.next

            if fnext:
                fnext.pre = None
                self._front.next = None

            self._front = fnext
        else:
            val: int = self._rear.val
            rpre: ListNode | None = self._rear.pre

            if rpre:
                rpre.next = None
                self._rear.pre = None

            self._rear = rpre

        self._size -= 1
        return val

    def pop_first(self) -> int:
        return self.pop(True)
    
    def pop_last(self) -> int:
        return self.pop(False)
    
    def peek_first(self) -> int:
        if self.is_empty():
            raise IndexError('out of index')
        
        return self._front.val
    
    def peek_last(self) -> int:
        if self.is_empty():
            raise IndexError('out of index')
        
        return self._rear.val
    
    def to_array(self) -> list:
        node = self._front
        res = [0] * self.size()

        for i in range(self.size()):
            res[i] = node.val
            node = node.next

        return res
    
```

```python
class ArrayDeque:
    def __init__(self, capacity: int):
        self._nums: list[int] = [0] * capacity
        self._front: int = 0
        self._size: int = 0

    def capacity(self) -> int:
        return len(self._nums)
    
    def size(self) -> int:
        return self._size
    
    def is_empty(self) -> bool:
        return self.size() == 0
    
    def index(self, i: int) -> int:
        return (i + self.capacity()) % self.capacity()
    
    def push_first(self, num: int):
        if self.size() == self.capacity():
            print('队列已满')
            return
        
        self._front = self.index(self._front - 1)
        self._nums[self._front] = num
        self._size += 1

    def push_last(self, num: int):
        if self.size() == self.capacity():
            print('队列已满')

        rear = self.index(self._front + self.size())

        self._nums[rear] = num
        self._size += 1

    def pop_first(self) -> int:
        num = self.peek_first()


    def peek_first(self) -> int:
        if self.is_empty():
            raise IndexError('双向队列为空')
        return self._nums[self._front]
    
    def peek_last(self) -> int:
        if self.is_empty():
            raise IndexError('双向队列为空')
        
        last = self.index(self._front + self._size - 1)
        
        return self._nums[last]
    
    def to_array(self) -> list[int]:
        res = []
        for i in range(self._size):
            res.append(self._nums[self.index(self._front + i)])

        return res
```
